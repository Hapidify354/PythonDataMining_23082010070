```python
import pandas as pd

df = pd.read_csv("Weather_Data_PreProcesing.csv")

# ===============================
# Feature Engineering
# ===============================

# 1. Pengkategorian Suhu
df["temp_category"] = pd.cut(
    df["temperature_celsius"],
    bins=[-50, 15, 30, 60],
    labels=["Cold", "Normal", "Hot"]
)

# 2. Levelling Kelembaban
df["humidity_level"] = pd.cut(
    df["humidity_percent"],
    bins=[0, 40, 70, 100],
    labels=["Low", "Medium", "High"]
)

# 3. Pengkategorian Tekanan Udara
df["pressure_category"] = pd.cut(
    df["pressure_hpa"],
    bins=[900, 1000, 1020, 1100],
    labels=["Low", "Normal", "High"]
)

# 4. Index Cuaca (kombinasi numerik)
df["weather_index"] = (
    df["temperature_celsius"] * 0.4 +
    df["humidity_percent"] * 0.4 +
    df["wind_speed_ms"] * 0.2
)

# ===============================
# Encoding kategori
# ===============================
df = pd.get_dummies(
    df,
    columns=["temp_category", "humidity_level", "pressure_category"],
    drop_first=True
)

# ===============================
# Save Output
# ===============================
df.to_csv("Weather_Data_FeatureEngineering.csv", index=False)

print("Feature Engineering selesai!")
print("Output: Weather_Data_FeatureEngineering.csv")
print(df.head())
```
