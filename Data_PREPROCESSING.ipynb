```python
import pandas as pd

# =========================
# CONFIGURATION
# =========================
INPUT_FILE = "Weather_Data_Mentah.csv"
OUTPUT_FILE = "Weather_Data_PreProcesing.csv"

# =========================
# LOAD DATA
# =========================
df = pd.read_csv(INPUT_FILE)

print("Data awal:")
print(df.head())
print("\nJumlah data awal:", len(df))

# =========================
# PREPROCESSING DATA
# =========================

# 1. Konversi datetime
df["datetime"] = pd.to_datetime(df["datetime"])

# 2. Cek missing value pada data
missing_values = df.isnull().sum()
print("\nMissing values:")
print(missing_values)
df = df.dropna()

# 3. Feature Engineering (waktu)
df["hour"] = df["datetime"].dt.hour
df["day"] = df["datetime"].dt.day
df["day_of_week"] = df["datetime"].dt.day_name()

# 4. Labeling hujan (classification target)
df["is_rain"] = df["weather_main"].apply(
    lambda x: 1 if x in ["Rain", "Thunderstorm", "Drizzle"] else 0
)

# 5. Encoding sederhana untuk weather_main (optional untuk EDA)
df["weather_encoded"] = df["weather_main"].astype("category").cat.codes

# 6. Cek Inkosistensi Data
print('\nCEK INKONSISTENSI DATA PADA KOLOM KATEGORIKAL:')
for column in df.select_dtypes(include='object').columns:
    print(f"\nKolom '{column}':")
    print(df[column].unique())
    print(f"Jumlah nilai unik: {df[column].nunique()}")

# =========================
# FINAL DATA SELECTION
# =========================
final_columns = [
    "datetime",
    "temperature_celsius",
    "humidity_percent",
    "pressure_hpa",
    "wind_speed_ms",
    "hour",
    "day",
    "day_of_week",
    "weather_main",
    "weather_encoded",
    "is_rain"
]

df_final = df[final_columns]

# =========================
# SIMPAN DATA YG SUDAH DI CLEANING
# =========================
df_final.to_csv(OUTPUT_FILE, index=False)

print("\nPreprocessing selesai!")
print(f"File disimpan sebagai: {OUTPUT_FILE}")
print("Jumlah data akhir:", len(df_final))
```
