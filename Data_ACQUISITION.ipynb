```python
import requests
import pandas as pd
from datetime import datetime

# =========================
# CONFIGURATION
# =========================
API_KEY = "e63a18bac54a7be52e4c3bd8df45315f"
CITY = "Surabaya,ID"
URL = "https://api.openweathermap.org/data/2.5/forecast"
OUTPUT_FILE = "Weather_Data_Mentah.csv"

# =========================
# API REQUEST
# =========================
params = {
    "q": CITY,
    "appid": API_KEY,
    "units": "metric"
}

response = requests.get(URL, params=params, timeout=10)
data = response.json()

# =========================
# DEBUG & VALIDASI
# =========================
print("Response code :", data.get("cod"))
print("Message       :", data.get("message"))

if str(data.get("cod")) != "200":
    raise RuntimeError(
        f"Request API gagal (cod={data.get('cod')}): {data.get('message')}"
    )

if "list" not in data:
    raise KeyError("Key 'list' tidak ditemukan pada response API.")

# =========================
# DATA EXTRACTION
# =========================
records = []

for item in data["list"]:
    records.append({
        "datetime": datetime.fromtimestamp(item["dt"]),
        "temperature_celsius": item["main"]["temp"],
        "humidity_percent": item["main"]["humidity"],
        "pressure_hpa": item["main"]["pressure"],
        "wind_speed_ms": item["wind"]["speed"],
        "weather_main": item["weather"][0]["main"],
        "weather_description": item["weather"][0]["description"]
    })

# =========================
# SAVE DATA
# =========================
df = pd.DataFrame(records)
df.to_csv(OUTPUT_FILE, index=False)

print(f"Data acquisition berhasil! Ukuran data : ", df.shape)
print(f"File disimpan sebagai: {OUTPUT_FILE}")
print(f"Total data rows: {len(df)}")
```
